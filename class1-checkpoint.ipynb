## Identify and handle missing values

### Convert the question mark to NaN which means not a number


```python
import numpy as np
```

```python
df2.replace("?", np.nan, inplace=True)
df2.head()
```

## Evaluate the missing data
## "False" represents there is data  and "True " represents there is no data

```python
missingdata=df2.isnull()
missingdata.head()
```

```python
missingnotnulldata=df2.notnull()
missingnotnulldata.head()
```

## Counting missing values in each column

```python
L = missingnotnulldata.columns.values.tolist()
for column in L:
    print(column)
    print(missingnotnulldata[column].value_counts())
```
